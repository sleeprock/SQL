#Группировка категориальных переменных

```python
df.groupby('education_level', dropna=False)['target'].agg(['count', 'mean'])
```

```python
#объединим в одну группу school и phd с мастерс
```

```python
#с помощью функии делаем новые категроии
def 
edu_group (x):
  if x in['High School', 'Primary School']:
    return 'School'
  if x in['High Masters', 'Phd']:
    return 'Masters_and_Phd'
  return x
```

```python
df['new_edu_lev'] = df['education_level'].apply(edu_group)
```

```python
df.groupby('new_edu_lev', dropna=False)['target'].agg(['count', 'mean'])
```
